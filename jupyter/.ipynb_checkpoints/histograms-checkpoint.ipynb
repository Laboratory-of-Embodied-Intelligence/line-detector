```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from copy import deepcopy as dp
from skimage import transform as tf
%matplotlib inline
```

```python
def imShow(img):
    plt.imshow(img, cmap="gray")
    plt.xticks([]), plt.yticks([])
    plt.show()
    
road = cv2.imread("curve_road.png")
road = cv2.resize(road, (640, 480))
imShow(road)

def showImg(images, size, labels=None, figsize=(14, 14)):
    
    fig = plt.figure(figsize=figsize)
    for i in range(1, size[0]*size[1] + 1):
        ax = fig.add_subplot(size[0], size[1], i)
        if labels:
            ax.set_title(labels[i - 1])
        plt.imshow(images[i-1], cmap="gray")
        plt.xticks([]), plt.yticks([])
    plt.show()
    
def getHist(img):
    r = cv2.calcHist([img], [0], None, [256],[0,256])
    g = cv2.calcHist([img], [1], None, [256],[0,256])
    b = cv2.calcHist([img], [2], None, [256],[0,256])
    
    return (r, g, b)

def showHist(hist, figsize=(14, 5)):
    fig = plt.figure(figsize=figsize)
    
    ax_1 = fig.add_subplot(1, 3, 1)
    ax_1.set_title("l channel")
    line1, = plt.plot(hist[0], color = (0, 0, 0))
    
    ax_2 = fig.add_subplot(1, 3, 2)
    ax_2.set_title("a channel")
    line2, = plt.plot(hist[1], color = (0, 0, 0))
    
    ax_3 = fig.add_subplot(1, 3, 3)
    ax_3.set_title("b channel")
    line1, = plt.plot(hist[2], color = (0, 0, 0))
    
    plt.show()
```

```python
lab_road = cv2.cvtColor(road, cv2.COLOR_BGR2LAB)
l, a, b = cv2.split(dp(lab_road))

hists = getHist( dp(lab_road) )

showImg([l, a, b], (1, 3), ("l", "a", "b"))
showHist(hists)
```

```python
def BHEMHB(hist, img):
    h, w = img.shape
    
    repeats = np.median( np.repeat( np.arange(0, 256), hist ) )
    
    n = len(repeats)
    norm_hist = hist / n
    
    median = np.median( repeats )
    
    min_val = repeats[0]
    max_val = repeats[-1]
    
    p_low = norm_hist[0:median]
    p_up = norm_hist[median:]
    
    p_low = np.log2(p_low + 1)
    p_up = np.log2(p_up + 1)
    
    cs_low = np.cumsum(p_low)
    cs_up = np.cumsum(p_up)
    
    for i in range(h):
        for j in range(w):
            if img[i, j] < median:
                img[i, j] = min_val + (median - 1 - min_val) * cs_low( img[i, j] )
            else:
                img[i, j] = median + (max_val - median) * cs_up( img[i, j] )
                
    return img
```
